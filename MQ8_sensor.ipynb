dividing Rs/R0 for air by Rs/R0 for H2@air yields **Rs_air/Rs_H2@air = (cH2/ppm)^1.45 * 10^-2.49**

image.png

this leads to **cH2/ppm = (Rs_air/Rs_H2@air*10^2.49)^1/1.45**    (https://en.wikipedia.org/wiki/Log–log_plot)

Rs_air determined in clean air; Rs_H2@air measured during experiment

**Rs_H2@air = Rs_air* (cH2/ppm)^-1.45*10^2.49**

**Vout = Vin * RL/(Rs+RL)**

```python
def Rs_air(Vout): # determine sensor resistance in clean air
  Vin = 5 # input voltage in V
  RL = 10e3 # load resistance in Ohm
  Rs_air = RL*(Vin/Vout-1)
  return Rs_air
```

```python
print('Rs_air = %.2f Ohm' %Rs_air(0.15))
```

```python
def cH2ppm(Rs_air,RL,Vout):
  Vin = 5 # input voltage in V
  RL = 10e3 # load resistance in Ohm
  Rs_air = 323e3 # sensor resistance in clean air in Ohm
  Rs_H2_air = RL*(Vin/Vout-1)
  cH2ppm = (Rs_air/Rs_H2_air*10**2.49)**1/1.45 # in ppm
  return cH2ppm
```

```python
print("cH2 = %.2f ppm" %cH2ppm(323e3,10e3,0.25))
```
