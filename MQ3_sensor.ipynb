dividing Rs/R0 for air by Rs/R0 for alcohol@air yields **Rs_air/Rs_Alc_air = (cAlc/mgl^-1)^0.665 * 10^+2.045**

image.png

this leads to **cAlc/mgl^-1 = (Rs_air/Rs_Alc_air*10^-2.045)^1/0.665**    (https://en.wikipedia.org/wiki/Log–log_plot)

Rs_air determined in clean air; Rs_Alc_air measured during alcohol sensing experiment

**Vout = Vin * RL/(Rs+RL)**

```python
def Rs_air(Vout): # determine sensor resistance in clean air
  Vin = 5 # input voltage in V
  RL = 200e3 # load resistance in Ohm
  Rs_air = RL*(Vin/Vout-1)
  return Rs_air
```

```python
print('Rs_air = %.2f Ohm' %Rs_air(0.1))
```

```python
def cAlc_mgl(Vout):
  Vin = 5 # input voltage in V
  RL = 200e3 # load resistance in Ohm
  Rs_air = 10e6 # sensor resistance in clean air in Ohm
  Rs_Alc_air = RL*(Vin/Vout-1)
  cAlc_mgl = (Rs_air/Rs_Alc_air*10**-2.045)**1/0.665 # in mgl^-1
  return cAlc_mgl
```

```python
print("cAlc = %.2f mgl^-1" %cAlc_mgl(0.25))
```
